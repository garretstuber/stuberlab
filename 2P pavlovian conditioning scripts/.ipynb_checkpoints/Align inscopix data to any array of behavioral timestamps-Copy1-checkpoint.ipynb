```python
#This script will take a raw numpy array generated by exporting the ROI output from Sima, and align all the trials
#together for each cell in the order that they were presented.  This output is then saved to a two new numpy arrays
#called CSplustrials and CSminustrials.

#you can then plot the output of any ROI aligned to the trial time or plot the mean of the cell.  
```

```python
#load dependants - 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import mpld3
mpld3.enable_notebook()
import pandas as pd
from pandas import Series, DataFrame, groupby, Panel
from seaborn import FacetGrid 
```

```python
#choose the directory of the data
%cd ~/Data/shanna
```

```python
%mkdir -p /Users/gstuber/Data/shanna/export
outputdir = '/Users/gstuber/Data/shanna/export'
```

```python
#load the main textfile of data
traces = np.loadtxt("traces.txt", delimiter=",")
```

```python
traces.shape
```

```python
#loads the timestamp arrays from csv files with the following variable names.
approach = np.loadtxt("approach.csv", delimiter=",")
manipulation = np.loadtxt("manipulation.csv", delimiter=",")
sniff = np.loadtxt("sniff.csv", delimiter=",")
```

```python
#this sets the array size for aligned events
numrois = traces.shape[1]-1
numsniffs = sniff.shape[0]
numapproach = approach.shape[0]
numtimesamples = 100
```

```python
numapproach
```

```python
#this generate a new array of zeros with trials on y, time on x, rois on z
alignsniff = np.zeros([numsniffs,numtimesamples,numrois])
#This will then clear all values to NaN
alignsniff[:] = np.NAN
```

```python
alignapproach = np.zeros([numapproach,numtimesamples,numrois])
#This will then clear all values to NaN
alignapproach[:] = np.NAN
```

```python
datatimestamps = traces[:,0]
```

```python
sniff[0]
```

```python
#Transposes the array signals, and assigns this to the array temp
#for i in range 0-100, the value of align gets added iteratively to each row (trial) from temp.  Running this will
#fill the align array with trials on y, trial time on x, and ROIs on z.

for i in range(0,numsniffs):
    tempindex = np.argwhere(datatimestamps - sniff[i] >0)[0]
    alignsniff[i,:,:] = traces[tempindex-50:tempindex+50,1:]
```

```python
#Transposes the array signals, and assigns this to the array temp
#for i in range 0-100, the value of align gets added iteratively to each row (trial) from temp.  Running this will
#fill the align array with trials on y, trial time on x, and ROIs on z.

for i in range(0,numapproach):
    tempindex2 = np.argwhere(datatimestamps - approach[i] >0)[0]
    alignapproach[i,:,:] = traces[tempindex2-50:tempindex2+50,1:]
```

```python
alignapproach
```

```python
#This plots the signal average for all rois across aligned to the sniff
for i in range(0,numrois):
    sns.tsplot(alignsniff[:,:,i])
    plt.savefig(outputdir+'/roi'+str(i)+'sniffaverage.tif')
    plt.gca().cla()
```

```python
#This plots the signal average for all rois across aligned to the approach
for i in range(0,numrois):
    sns.tsplot(alignapproach[:,:,i])
    plt.savefig(outputdir+'/aroi'+str(i)+'approachaverage.tif')
    plt.gca().cla()
```
